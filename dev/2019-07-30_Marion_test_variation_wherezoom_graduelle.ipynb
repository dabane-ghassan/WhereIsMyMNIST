## Génération de datasets graduellement "zoomés"

### Datasets avec transformdico

```python
list_transform = []
```

```python
for nb_diziemes in range(10, 0, -1):
    zoomW = nb_diziemes / 10 * args.N_Y
    transformation = transforms.Compose([
            WhereZoom(args, zoomW),
            WhereGrey(args),
            RetinaWhiten(args, resize=False), 
            TransformDico(retina)])
    list_transform.append(transformation)
```

```python
for i in range(len(list_transform)):
    what_gender_trainer = WhatGenderTrainer(args, retina, transform=list_transform[i])
```
